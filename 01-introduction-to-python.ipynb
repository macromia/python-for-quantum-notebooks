## 2. Variables and Types

A **variable** is a name that refers to a value stored in memory. You create one with `=` (assignment, not equality).

```python
n_qubits = 5
energy = -13.6
element = "hydrogen"
is_entangled = True

print(n_qubits)
print(energy)
print(element)
print(is_entangled)
```

## 3. Arithmetic

Python supports the full set of arithmetic operations:

```python
a = 10
b = 3

print(a + b)   # Addition:        13
print(a - b)   # Subtraction:      7
print(a * b)   # Multiplication:  30
print(a / b)   # Division:        3.333...
print(a // b)  # Floor division:   3
print(a % b)   # Modulo:           1
print(a ** b)  # Exponentiation: 1000
```

> **Key distinction:** `/` always returns a `float`. `//` returns the quotient as an integer (the remainder is discarded). In quantum computing, floor division appears when computing bit-widths: a system of `n` states needs `n // 2 + 1` bits to index them.

```python
# Operator precedence: exponentiation before multiplication, before addition
print(2 ** 3 + 1)     # 9  (exponent first)
print(2 ** (3 + 1))   # 16 (parentheses override)
```

```python
# f-strings: embed expressions directly in text
n = 3
print(f"A {n}-qubit system has {2**n} basis states.")
```

## 4. The `complex` Type

Python has a built-in type for complex numbers, a first-class part of the language, not an add-on library. Quantum states are described by complex amplitudes, so this type will appear everywhere in this course.

In Python, the imaginary unit is written **`j`** (not `i`). A complex number is written as:

```python
# Two equivalent ways to write the same complex number
z1 = 3 + 4j
z2 = complex(3, 4)

print(z1)            # (3+4j)
print(z2)            # (3+4j)
print(z1 == z2)      # True
print(type(z1))      # <class 'complex'>
```

### Real and Imaginary Parts

Access the components directly as attributes. Both are returned as `float`.

```python
z = 3 + 4j
print(z.real)   # 3.0
print(z.imag)   # 4.0
```

### The Modulus: `abs()`

The **modulus** of $z = a + bj$ is $|z| = \sqrt{a^2 + b^2}$, the distance from the origin in the complex plane. In quantum mechanics, the modulus squared of an amplitude gives a probability.

```python
z = 3 + 4j
modulus = abs(z)
print(f"|z| = {modulus}")           # 5.0
print(f"|z|² = {modulus**2}")       # 25.0  (this is the probability)

# Verify: sqrt(3² + 4²) = sqrt(9 + 16) = sqrt(25) = 5
import math
print(math.sqrt(3**2 + 4**2))       # 5.0
```

### The Complex Conjugate: `.conjugate()`

The **conjugate** of $z = a + bj$ is $\bar{z} = a - bj$: same real part, negated imaginary part. The product $z \cdot \bar{z} = |z|^2$ is always a non-negative real number.

```python
z = 3 + 4j
z_conj = z.conjugate()

print(f"z      = {z}")
print(f"z*     = {z_conj}")
print(f"z · z* = {z * z_conj}")   # (25+0j) always real and non-negative
```

### Arithmetic with Complex Numbers

All standard operators work on complex numbers. Python handles the algebra automatically.

```python
z1 = 1 + 2j
z2 = 3 - 1j

print(f"z1 + z2 = {z1 + z2}")   # (4+1j)
print(f"z1 - z2 = {z1 - z2}")   # (-2+3j)
print(f"z1 * z2 = {z1 * z2}")   # (5+5j)
print(f"z1 / z2 = {z1 / z2}")   # (0.1+0.7j)
```

### Quantum Amplitude Example

A single qubit in superposition has a state:

$$|\psi\rangle = \alpha|0\rangle + \beta|1\rangle$$

where $\alpha$ and $\beta$ are complex amplitudes satisfying $|\alpha|^2 + |\beta|^2 = 1$.

Let's verify a valid qubit state:

```python
import math

# A qubit in an equal superposition with a phase
alpha = 1/math.sqrt(2) + 0j          # amplitude for |0⟩
beta  = 0 + 1j / math.sqrt(2)        # amplitude for |1⟩  (pure imaginary)

prob_0 = abs(alpha)**2
prob_1 = abs(beta)**2

print(f"α = {alpha:.4f}")
print(f"β = {beta:.4f}")
print(f"|α|² = {prob_0:.4f}   (probability of measuring |0⟩)")
print(f"|β|² = {prob_1:.4f}   (probability of measuring |1⟩)")
print(f"|α|² + |β|² = {prob_0 + prob_1:.4f}   (must equal 1)")
```

## 5. Type Conversion

Python allows explicit conversion between types using **casting functions**.

```python
print(int(3.9))      # 3     truncates toward zero, does NOT round
print(int(-3.9))     # -3    truncates toward zero
print(float(7))      # 7.0
print(str(42))       # '42'
print(bool(0))       # False
print(bool(99))      # True; anything nonzero is True
```

```python
# Python promotes types automatically in mixed arithmetic
result = 3 + 1.5       # int + float → float
print(result)          # 4.5
print(type(result))    # <class 'float'>
```

```python
# Any real number can become complex
x = 5.0
z = complex(x)         # (5+0j)
print(z)
print(type(z))

# Going the other direction loses information
w = 7 + 0j
print(w.real)          # 7.0 (extract the real part manually)
```

> **Truncation vs rounding:** `int(3.9)` gives `3`, not `4`. Use `round()` when you want the nearest integer.

```python
print(int(3.9))     # 3 (truncation)
print(round(3.9))   # 4 (rounding)
```
