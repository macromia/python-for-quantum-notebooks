## 1. `if`, `elif`, and `else`

An `if` statement evaluates a condition and runs a block of code only when that condition is `True`.

> **Syntax rule:** The body of every `if`, `elif`, and `else` block must be indented (four spaces by convention). Indentation is not style in Python; it is syntax.

````python
measurement = 1   # result of measuring a qubit: 0 or 1

if measurement == 1:
    print("Qubit collapsed to |1⟩")
else:
    print("Qubit collapsed to |0⟩")
````

````python
# elif checks additional conditions in sequence
energy_level = 3

if energy_level == 0:
    print("Ground state")
elif energy_level == 1:
    print("First excited state")
elif energy_level == 2:
    print("Second excited state")
else:
    print(f"Higher excited state: n = {energy_level}")
````

Combine conditions with `and`, `or`, and `not`:

### Quantum Example: Classifying a Measurement Outcome

Given the probability of measuring $|1\rangle$, classify the qubit state into qualitative categories.

````python
def classify_outcome(prob_one):
    """Classify a qubit state by its probability of measuring |1⟩."""
    if prob_one > 0.99:
        print("Effectively certain: |1⟩")
    elif prob_one > 0.5:
        print(f"More likely |1⟩, with probability {prob_one:.2f}")
    elif prob_one == 0.5:
        print("Maximum superposition: equal probability")
    elif prob_one > 0.01:
        print(f"More likely |0⟩, with probability {1 - prob_one:.2f}")
    else:
        print("Effectively certain: |0⟩")

classify_outcome(0.72)
classify_outcome(0.5)
classify_outcome(0.03)
````

## 2. `for` Loops

A `for` loop repeats a block of code once for each item in a sequence.

````python
qubits = ["q0", "q1", "q2"]

for qubit in qubits:
    print(f"Applying Hadamard gate to {qubit}")
````

### `range()`: Looping a Fixed Number of Times

`range(n)` produces the integers from `0` up to but not including `n`.

````python
# Basic range
for i in range(5):
    print(i, end=" ")   # prints: 0 1 2 3 4
print()

# range with start, stop, step
for i in range(2, 10, 2):
    print(i, end=" ")   # prints: 2 4 6 8
print()

# Counting down
for i in range(4, -1, -1):
    print(i, end=" ")   # prints: 4 3 2 1 0
print()
````

### Accumulating Values

A common pattern: start with an initial value and update it on each iteration.

````python
outcomes = [0, 1, 0, 0, 1, 1, 0, 1]
total_ones = 0

for outcome in outcomes:
    total_ones += outcome

print(f"Measured |1⟩ {total_ones} times out of {len(outcomes)}")
print(f"Estimated probability: {total_ones / len(outcomes):.3f}")
````

### `enumerate()`: Index and Value Together

When you need both the index and the value during a loop, use `enumerate()`.

````python
basis_states = ["|00⟩", "|01⟩", "|10⟩", "|11⟩"]
amplitudes   = [0.5, 0.5, 0.5, 0.5]

for i, state in enumerate(basis_states):
    prob = amplitudes[i]**2
    print(f"State {state}: amplitude = {amplitudes[i]:.2f}, probability = {prob:.2f}")
````

### Quantum Example: Probability Distribution Over Basis States

Compute and display the probability distribution for an equal superposition of all $n$-qubit basis states.

````python
import math

n_qubits = 3
n_states = 2**n_qubits
amplitude = 1 / math.sqrt(n_states)

print(f"{'State':<10} {'Amplitude':>12} {'Probability':>14}")
print("-" * 38)

total_prob = 0.0
for k in range(n_states):
    binary = format(k, f"0{n_qubits}b")
    prob = amplitude**2
    total_prob += prob
    print(f"|{binary}⟩ {amplitude:>14.4f} {prob:>14.4f}")

print("-" * 38)
print(f"{'Total':>25} {total_prob:>14.4f}")
````

## 3. `while` Loops

A `while` loop repeats a block of code as long as a condition remains `True`. Unlike a `for` loop, it does not iterate over a fixed sequence; it keeps going until something inside the loop changes the condition.

> **Warning:** Always make sure the loop body contains something that will eventually make the condition `False`. A loop that never terminates runs forever. In Colab, stop a runaway cell with the stop button or **Ctrl + M, I**.

### Quantum Example: Repeat-Until-Success

Some quantum protocols are probabilistic: attempt an operation, check whether it succeeded, and repeat if it did not. This pattern is common in linear optical quantum computing and quantum communication.

````python
import random
random.seed(42)   # fix seed for reproducibility

attempts = 0
success = False

while not success:
    attempts += 1
    outcome = random.random()   # uniform random number in [0, 1)

    if outcome < 0.25:          # success probability: 25%
        success = True
        print(f"Success on attempt {attempts}")
    else:
        print(f"Attempt {attempts}: failed, retrying...")
````

### `break` and `continue`

- `break` exits the loop immediately.
- `continue` skips the rest of the current iteration and returns to the condition check.

````python
# Find the first qubit with a measurement error
measurements = [0, 0, 1, 0, 1, 1, 0]
expected     = [0, 0, 0, 0, 1, 1, 0]

for i in range(len(measurements)):
    if measurements[i] == expected[i]:
        continue   # correct outcome; skip to the next
    print(f"Error at qubit {i}: measured {measurements[i]}, expected {expected[i]}")
    break          # report only the first error
````

### Convergence Loop

A common pattern in variational quantum algorithms: iterate until the update falls below a threshold.

````python
import random
random.seed(7)

energy = 10.0
threshold = 0.001
step = 0

while True:
    step += 1
    update = random.uniform(-0.5, 0.5) * energy * 0.1
    energy += update

    if step <= 5 or abs(update) < threshold:   # print first 5 steps and the final one
        print(f"Step {step:3d}: energy = {energy:.6f}, update = {update:.6f}")

    if abs(update) < threshold:
        print(f"\nConverged after {step} steps. Final energy: {energy:.6f}")
        break
````

## 4. List Comprehensions

A **list comprehension** is a concise way to build a list by applying an expression to every element of a sequence, with an optional filter condition.

```python
[expression for item in sequence]
[expression for item in sequence if condition]
```

````python
# Using a for loop
squares_loop = []
for n in range(5):
    squares_loop.append(n**2)

# Using a list comprehension
squares_comp = [n**2 for n in range(5)]

print(squares_loop)   # [0, 1, 4, 9, 16]
print(squares_comp)   # [0, 1, 4, 9, 16]
print(squares_loop == squares_comp)   # True
````

### With a Filter Condition

Add an `if` clause to include only items satisfying a condition.

````python
# All 3-qubit basis states whose first bit is 1
n_qubits = 3
all_states = [format(k, f"0{n_qubits}b") for k in range(2**n_qubits)]
print("All states:    ", all_states)

starts_with_one = [s for s in all_states if s[0] == "1"]
print("First bit = 1: ", starts_with_one)
````

### Quantum Applications

List comprehensions are particularly useful for generating state labels and computing probabilities from amplitudes.

````python
# Generate all basis state labels for a 2-qubit system
n_qubits = 2
basis = [f"|{format(k, f'0{n_qubits}b')}⟩" for k in range(2**n_qubits)]
print("Basis states:", basis)

# Compute probabilities from amplitudes
amplitudes = [1/2, 1/2, 1/2, 1/2]
probs = [abs(a)**2 for a in amplitudes]
print("Probabilities:", probs)
print(f"Total: {sum(probs):.1f}")

# Keep only states with probability above a threshold
threshold = 0.1
significant = [(basis[i], probs[i]) for i in range(len(probs)) if probs[i] > threshold]
print("Significant states:", significant)
````

## Summary

| Tool | Use case | Key syntax |
|------|----------|------------|
| `if / elif / else` | Branch on a condition | `if x > 0:` / `elif x == 0:` / `else:` |
| `for` | Iterate over a known sequence | `for item in sequence:` |
| `range()` | Generate integers for a `for` loop | `range(n)`, `range(start, stop, step)` |
| `enumerate()` | Index and value together | `for i, v in enumerate(seq):` |
| `while` | Repeat until a condition changes | `while condition:` |
| `break` / `continue` | Exit or skip inside a loop | `break` / `continue` |
| List comprehension | Build a list concisely | `[expr for x in seq if cond]` |

**What comes next:** Control flow gives you branching and repetition. The next lesson introduces **functions**, which package logic into reusable, named units. Functions are how you avoid repeating yourself and how quantum simulations are structured as modular, testable pieces.
