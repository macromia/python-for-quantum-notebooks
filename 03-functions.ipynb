## 1. Defining and Calling a Function

A function is defined with the `def` keyword. Defining it only creates it; calling it runs the body.

````python
def greet():
    print("Welcome to quantum computing.")

greet()   # call the function
````

### Parameters and Arguments

**Parameters** are the input names listed in the `def` line. **Arguments** are the actual values passed when calling the function.

````python
def apply_gate(gate_name, qubit):
    print(f"Applying {gate_name} gate to qubit {qubit}")

apply_gate("Hadamard", "q0")
apply_gate("CNOT", "q1")
````

## 2. Return Values

`return` sends a value back to the caller and exits the function immediately.

````python
def modulus(z):
    """Return the modulus of a complex number z."""
    return abs(z)

m = modulus(3 + 4j)
print(m)              # 5.0

# Use the return value directly in an expression
prob = modulus(1 + 1j)**2
print(f"Probability: {prob:.4f}")   # 2.0000
````

### Returning Multiple Values

Separate return values with commas. Python packs them into a tuple; you unpack it on the receiving side.

````python
import math

def polar_form(z):
    """Return the modulus and phase angle of a complex number."""
    r = abs(z)
    theta = math.atan2(z.imag, z.real)
    return r, theta

r, theta = polar_form(1 + 1j)
print(f"r     = {r:.4f}")
print(f"theta = {theta:.4f} rad  ({math.degrees(theta):.1f} degrees)")
````

### Functions Without a Return Value

A function with no `return` statement returns `None`. Functions that only produce output (printing, plotting) often return `None` intentionally.

````python
def print_state(label, amplitude):
    prob = abs(amplitude)**2
    print(f"State {label}: amplitude = {amplitude}, probability = {prob:.4f}")

result = print_state("|0⟩", 0.6 + 0.8j)
print(f"Return value: {result}")   # None
````

## 3. Default Parameter Values and Keyword Arguments

A default value makes a parameter optional. Parameters with defaults must come after parameters without defaults.

````python
import random

def measure(prob_one, n_shots=1000, seed=None):
    """Simulate n_shots measurements. Returns the empirical probability of |1⟩."""
    if seed is not None:
        random.seed(seed)
    outcomes = [1 if random.random() < prob_one else 0 for _ in range(n_shots)]
    return sum(outcomes) / n_shots

print(measure(0.7))                          # uses defaults: n_shots=1000, seed=None
print(measure(0.7, 100))                     # n_shots=100
print(measure(0.7, n_shots=500, seed=42))    # keyword arguments, reproducible
````

### Keyword Arguments

Naming arguments explicitly makes calls self-documenting and allows any order.

````python
def state_info(label, amplitude, n_qubits=1):
    """Print information about a basis state."""
    prob = abs(amplitude)**2
    print(f"[{n_qubits}-qubit system] {label}: |α| = {abs(amplitude):.3f}, P = {prob:.3f}")

state_info("|0⟩", 0.6 + 0.8j)
state_info("|01⟩", 0.5 + 0j, n_qubits=2)
state_info(n_qubits=3, label="|010⟩", amplitude=0.5)   # any order with keywords
````

## 4. Scope

A variable created inside a function is **local**: it exists only within that function and disappears when the function returns.

````python
def compute_probability(amplitude):
    prob = abs(amplitude)**2   # 'prob' is local to this function
    return prob

p = compute_probability(0.6 + 0.8j)
print(f"p = {p}")   # 1.0

# Uncomment the line below to see the NameError:
# print(prob)   # NameError: name 'prob' is not defined
````

### Global Variables

Variables defined at the top level of a notebook are **global** and readable inside functions. By convention, global constants are written in uppercase.

````python
PLANCK = 6.626e-34   # J·s
SPEED_OF_LIGHT = 2.998e8   # m/s

def photon_energy(frequency):
    """Return photon energy E = h·f in joules."""
    return PLANCK * frequency

def photon_wavelength(frequency):
    """Return photon wavelength λ = c/f in meters."""
    return SPEED_OF_LIGHT / frequency

f = 5e14   # visible light: ~600 nm, orange
print(f"Energy:     {photon_energy(f):.3e} J")
print(f"Wavelength: {photon_wavelength(f)*1e9:.1f} nm")
````

### Why Scope Matters

Scope isolation means functions cannot accidentally modify variables in the rest of the program. Each function call gets its own workspace. This is what makes functions safe to reuse and safe to test independently.

````python
x = 100   # global variable

def double(value):
    x = value * 2   # this 'x' is local; does NOT touch the global x
    return x

print(double(5))   # 10
print(x)           # 100 — the global x is unchanged
````

## 5. Docstrings

A **docstring** is a string literal placed immediately after the `def` line. It documents the function and is accessible via `help()`.

````python
def inner_product(bra, ket):
    """
    Compute the inner product <bra|ket> for two quantum state vectors.

    Parameters
    ----------
    bra : list of complex
        The bra vector (conjugated internally).
    ket : list of complex
        The ket vector.

    Returns
    -------
    complex
        The inner product, a single complex number.
    """
    if len(bra) != len(ket):
        raise ValueError("Vectors must have the same length.")
    return sum(b.conjugate() * k for b, k in zip(bra, ket))

help(inner_product)
````

````python
import math

# Equal superposition state: |+⟩ = (1/√2)|0⟩ + (1/√2)|1⟩
plus  = [1/math.sqrt(2), 1/math.sqrt(2)]
minus = [1/math.sqrt(2), -1/math.sqrt(2)]
zero  = [1.0, 0.0]

print(f"<+|+⟩  = {inner_product(plus, plus):.4f}")    # 1 (normalized)
print(f"<+|->  = {inner_product(plus, minus):.4f}")   # 0 (orthogonal)
print(f"<0|+⟩  = {inner_product(zero, plus):.4f}")    # 1/√2
````

## 6. Putting It Together: A Small Quantum Toolkit

Small, single-purpose functions compose into larger simulations. Each can be tested independently.

````python
import math

def normalize(amplitudes):
    """Return a normalized copy of a complex amplitude list."""
    norm = math.sqrt(sum(abs(a)**2 for a in amplitudes))
    if norm == 0:
        raise ValueError("Cannot normalize the zero vector.")
    return [a / norm for a in amplitudes]

def probabilities(amplitudes):
    """Return the probability for each amplitude in the list."""
    return [abs(a)**2 for a in amplitudes]

def measure_once(probs):
    """Sample one basis state index according to the probability distribution."""
    import random
    r = random.random()
    cumulative = 0.0
    for i, p in enumerate(probs):
        cumulative += p
        if r < cumulative:
            return i
    return len(probs) - 1
````

````python
import random
random.seed(42)

# A non-normalized state: amplitudes proportional to (3, 4)
raw   = [3 + 0j, 4 + 0j]
state = normalize(raw)
probs = probabilities(state)

print(f"Raw amplitudes:  {raw}")
print(f"Normalized:      {[f'{a:.3f}' for a in state]}")
print(f"Probabilities:   {[f'{p:.3f}' for p in probs]}")
print(f"Sum of probs:    {sum(probs):.4f}")

# Run 10 measurements
print("\n10 measurements:")
for _ in range(10):
    outcome = measure_once(probs)
    print(f"  |{outcome}⟩", end="")
print()
````

## Summary

| Concept | Key syntax | Notes |
|---------|-----------|-------|
| Define a function | `def name(params):` | Body must be indented |
| Call a function | `name(args)` | Runs the body with the given arguments |
| Return a value | `return value` | Exits immediately; no `return` gives `None` |
| Multiple returns | `return a, b` | Packed as a tuple; unpack with `x, y = ...` |
| Default parameter | `def f(x, n=10):` | `n` is optional; defaults used when not passed |
| Keyword argument | `f(n=10, x=5)` | Named explicitly; any order allowed |
| Local scope | Variable inside `def` | Invisible outside; disappears on return |
| Global constant | Defined outside `def` | Readable inside functions; use uppercase name |
| Docstring | String after `def` line | Documents the function; shown by `help()` |
