# Data Structures

A quantum simulation is not just computation; it is also bookkeeping. You need somewhere to store a list of amplitudes, somewhere to map qubit labels to outcomes, somewhere to track which syndromes have fired. Python's four core data structures handle these different organizational needs.

This notebook accompanies **Lesson 4** of *Python Programming for Quantum Technology I*. Topics covered:
- `list`: ordered, mutable sequences
- `tuple`: ordered, immutable records
- `dict`: key-value storage with fast lookup
- `set`: unordered unique elements and set algebra

---
## 1. Lists

A **list** is an ordered, mutable sequence. Ordered means items have a fixed position; mutable means you can change them after creation.

### Indexing

Access items by position. Indexing starts at zero; negative indices count from the end.

````python
qubits = ["q0", "q1", "q2", "q3"]

print(qubits[0])    # "q0"  first item
print(qubits[-1])   # "q3"  last item
print(qubits[2])    # "q2"
````

### Slicing

Extract a sub-list with `start:stop:step`. The stop index is **excluded**.

````python
outcomes = [0, 1, 0, 0, 1, 1, 0, 1]

print(outcomes[2:5])    # [0, 0, 1]         indices 2, 3, 4
print(outcomes[:3])     # [0, 1, 0]         from the start
print(outcomes[5:])     # [1, 0, 1]         to the end
print(outcomes[::2])    # [0, 0, 1, 0]      every other item
print(outcomes[::-1])   # [1, 0, 1, 1, 0, 0, 1, 0]  reversed
````

### Mutability and Common Methods

````python
# Lists can be changed in place
state = [1.0, 0.0]   # |0⟩ amplitude vector
state[0] = 0.0
state[1] = 1.0       # now represents |1⟩
print(state)         # [0.0, 1.0]
````

````python
results = []

results.append(0)           # add one item to the end
results.append(1)
results.extend([0, 0, 1])   # add all items from another list
print(results)              # [0, 1, 0, 0, 1]

results.pop()               # remove and return the last item
print(results)              # [0, 1, 0, 0]

print(sorted(results))      # [0, 0, 0, 1]  new sorted list; original unchanged
print(results)              # [0, 1, 0, 0]  still the original order

print(len(results))         # 4
print(results.count(0))     # 3
````

### Nested Lists: Matrices

A list of lists represents a matrix. You will use this pattern before moving to NumPy.

````python
import math

# Hadamard gate (unnormalized)
H_unnorm = [
    [1,  1],
    [1, -1]
]

# Hadamard gate (normalized)
s = 1 / math.sqrt(2)
H = [
    [s,  s],
    [s, -s]
]

print("H[0]   =", H[0])      # first row
print("H[1][0] =", H[1][0])  # row 1, column 0
````

## 2. Tuples

A **tuple** is an ordered, immutable sequence written with parentheses. Once created, its contents cannot be changed.

````python
state_zero = (1.0, 0.0)   # amplitude vector for |0⟩
state_one  = (0.0, 1.0)

print(state_zero[0])   # 1.0  (indexing works like lists)
print(type(state_zero))

# Attempting to modify a tuple raises a TypeError:
try:
    state_zero[0] = 0.5
except TypeError as e:
    print(f"TypeError: {e}")
````

### Tuple Unpacking

Assign each element of a tuple to a separate variable in one line.

````python
result = ("q2", 1, 0.87)   # (qubit label, outcome, confidence)

label, outcome, confidence = result
print(f"Qubit {label} measured {outcome} with confidence {confidence:.2f}")
````

````python
# Single-element tuples require a trailing comma
not_a_tuple = (42)
is_a_tuple  = (42,)

print(type(not_a_tuple))   # <class 'int'>
print(type(is_a_tuple))    # <class 'tuple'>
````

### When to Use a Tuple

Use a tuple when the collection of values should not change: fixed system dimensions, measurement records, physical constants grouped together, or function return values.

````python
import math

# Fixed physical constants
CONSTANTS = {
    "planck":  (6.626e-34, "J·s"),
    "hbar":    (1.055e-34, "J·s"),
    "c":       (2.998e8,   "m/s"),
    "k_B":     (1.381e-23, "J/K"),
}

for name, (value, unit) in CONSTANTS.items():
    print(f"{name:8s} = {value:.3e} {unit}")
````

## 3. Dictionaries

A **dictionary** stores key-value pairs. Given a key, it retrieves the associated value in constant time. Keys must be unique and immutable.

````python
# Map basis state labels to their amplitudes
state = {
    "|00⟩": 0.5 + 0j,
    "|01⟩": 0.5 + 0j,
    "|10⟩": 0.5 + 0j,
    "|11⟩": 0.5 + 0j,
}

print(state["|00⟩"])              # (0.5+0j)
print(state.get("|10⟩"))          # (0.5+0j)
print(state.get("|99⟩", "N/A"))   # "N/A"  (key absent; return default)
````

### Adding and Updating Entries

````python
gate_counts = {}

gate_counts["H"]    = 3     # add new key
gate_counts["CNOT"] = 2
gate_counts["H"]   += 1    # update existing key

print(gate_counts)   # {'H': 4, 'CNOT': 2}
````

### Iterating Over a Dictionary

Three views: `.keys()`, `.values()`, and `.items()`. The `.items()` view is the most common.

````python
outcomes = {0: 1, 1: 0, 2: 1, 3: 0}   # qubit index -> measurement outcome

print("Keys:  ", list(outcomes.keys()))
print("Values:", list(outcomes.values()))
print()

for qubit, result in outcomes.items():
    print(f"q{qubit} measured |{result}⟩")
````

````python
# Membership test: check if a key exists
if "|00⟩" in state:
    print(f"Amplitude of |00⟩: {state['|00⟩']}")

if "|99⟩" not in state:
    print("|99⟩ is not in this state.")
````

### Dictionary Comprehensions

Like list comprehensions, but producing key-value pairs.

````python
import math

n_qubits = 3
n_states = 2**n_qubits
amplitude = 1 / math.sqrt(n_states)

# Equal superposition state as a dictionary
superposition = {
    format(k, f"0{n_qubits}b"): amplitude
    for k in range(n_states)
}

for label, amp in superposition.items():
    print(f"|{label}⟩: amplitude = {amp:.4f}, probability = {amp**2:.4f}")
````

### Quantum Application: A Gate Library

````python
import math

s = 1 / math.sqrt(2)

gates = {
    "I": [[1, 0], [0,  1]],
    "X": [[0, 1], [1,  0]],
    "Z": [[1, 0], [0, -1]],
    "H": [[s, s], [s, -s]],
}

def apply_gate(gate_name, state_vector):
    """Apply a named 2x2 gate to a 2-element state vector."""
    if gate_name not in gates:
        raise ValueError(f"Unknown gate: {gate_name}")
    matrix = gates[gate_name]
    a, b = state_vector
    new_a = matrix[0][0] * a + matrix[0][1] * b
    new_b = matrix[1][0] * a + matrix[1][1] * b
    return [new_a, new_b]

# Apply Hadamard to |0⟩ = [1, 0]
ket_0 = [1.0, 0.0]
plus_state = apply_gate("H", ket_0)
print(f"H|0⟩ = [{plus_state[0]:.4f}, {plus_state[1]:.4f}]")

# Apply X (NOT) gate to |0⟩
ket_1 = apply_gate("X", ket_0)
print(f"X|0⟩ = {ket_1}")
````

## 4. Sets

A **set** is an unordered collection of unique elements. Duplicates are silently ignored. Sets support fast membership testing and standard mathematical set operations.

````python
# Create a set from a list (duplicates removed automatically)
from_list = set([0, 1, 0, 1, 0, 1])
print(from_list)           # {0, 1}

measured  = {"q0", "q2", "q3"}
all_qubits = {"q0", "q1", "q2", "q3"}

# An empty set must use set(), not {}
empty_set = set()
print(type(empty_set))     # <class 'set'>
print(type({}))            # <class 'dict'>  (not a set!)
````

### Membership Testing

````python
active_qubits = {"q0", "q1", "q4"}

for q in ["q0", "q1", "q2", "q3", "q4"]:
    status = "active" if q in active_qubits else "idle"
    print(f"{q}: {status}")
````

### Set Operations

| Operation | Operator | Meaning |
|-----------|----------|----------|
| Union | `A \| B` | All elements from either A or B |
| Intersection | `A & B` | Elements in both A and B |
| Difference | `A - B` | Elements in A but not in B |
| Symmetric difference | `A ^ B` | Elements in A or B but not both |

````python
error_set_a = {"q1", "q3", "q5"}
error_set_b = {"q3", "q5", "q7"}

print(f"Union:                {error_set_a | error_set_b}")
print(f"Intersection:         {error_set_a & error_set_b}")
print(f"In A but not B:       {error_set_a - error_set_b}")
print(f"In B but not A:       {error_set_b - error_set_a}")
print(f"Symmetric difference: {error_set_a ^ error_set_b}")
````

### Quantum Application: Syndrome Tracking

In quantum error correction, each measurement round produces a set of triggered stabilizer checks. Comparing consecutive rounds reveals new errors and corrections.

````python
round_1 = {"s0", "s2"}           # syndromes triggered in round 1
round_2 = {"s2", "s4"}           # syndromes triggered in round 2

new_syndromes = round_2 - round_1   # appeared: new error event
cleared       = round_1 - round_2   # disappeared: error corrected
persistent    = round_1 & round_2   # same error still active

print(f"New syndromes:  {new_syndromes}")
print(f"Cleared:        {cleared}")
print(f"Persistent:     {persistent}")
````

## Summary

| Structure | Ordered | Mutable | Duplicates | Access | Best for |
|-----------|---------|---------|------------|--------|----------|
| `list` | Yes | Yes | Yes | By index | Sequences that grow or change |
| `tuple` | Yes | No | Yes | By index | Fixed records, function returns |
| `dict` | Yes | Yes | Keys: No | By key | Labelled data, fast lookup |
| `set` | No | Yes | No | N/A | Membership tests, set algebra |
